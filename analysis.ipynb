```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('matches.csv')

data = data.drop(columns=['Unnamed: 0'])
data['Home'] = data['Home'].str.strip()
data['Away'] = data['Away'].str.strip()

data['Score'] = data['Score'].str.replace(r'\s*\(.*?\)', '', regex=True)

data[['Home Goals', 'Away Goals']] = data['Score'].str.split('–', expand=True).astype(int)

def get_winner(row):
    if row['Home Goals'] > row['Away Goals']:
        return row['Home']
    elif row['Home Goals'] < row['Away Goals']:
        return row['Away']
    else:
        return 'Draw'

data['Winner'] = data.apply(get_winner, axis=1)


data
```

```python

match_wins = data['Winner'].value_counts()
wins = match_wins.to_frame('wins')
wins.to_csv('wins.csv')
```

```python
home_goals = data[['Home', 'Home Goals']].rename(columns={'Home': 'Team', 'Home Goals': 'Goals'})

away_goals = data[['Away', 'Away Goals']].rename(columns={'Away': 'Team', 'Away Goals': 'Goals'})

all_goals = pd.concat([home_goals, away_goals])

total_goals = all_goals.groupby('Team')['Goals'].sum().sort_values(ascending=False)

print(total_goals)

total_goals.to_csv('total-goals.csv')
```
